```
## import necessary libraries
from sklearn import datasets
```

```
## load iris dataset
iris_data = datasets.load_iris()
```

```
type(iris_data)
```

```
len(iris_data)
```

```
## extract features and target variables from iris dataset
X = iris_data.data
y = iris_data.target
```

```
print(f"Iris Target names {iris_data.target_names}")
```

```
## split dataset into train & test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
print(f"Lenght of train and test datasets :- \n X_train : {len(X_train)} \n y_train : {len(y_train)} \n X_test : {len(X_test)} \n y_test : {len(y_test)}")
```

```
## import and initialize decision tree a model
from sklearn import tree
tree_clf = tree.DecisionTreeClassifier()

## train decision tree model
tree_clf.fit(X_train, y_train)

## predict and evaluate decision tree trained model on test dataset
y_predict = tree_clf.predict(X_test)

## accuracy score 
from sklearn.metrics import accuracy_score, classification_report
accuracy_val = accuracy_score(y_test, y_predict)
classification_report_val = classification_report(y_test, y_predict)
print(f"Accuracy value : {accuracy_val}")
print(f"Classification Report : \n {classification_report_val}")
```

```
## import and initialize KNN  model
from sklearn import neighbors
knn_clf = neighbors.KNeighborsClassifier()

## train decision tree model
knn_clf.fit(X_train, y_train)

## predict and evaluate decision tree trained model on test dataset
y_predict = knn_clf.predict(X_test)

## accuracy score 
from sklearn.metrics import accuracy_score, classification_report
accuracy_val = accuracy_score(y_test, y_predict)
classification_report_val = classification_report(y_test, y_predict)
print(f"Accuracy value : {accuracy_val}")
print(f"Classification Report : \n {classification_report_val}")
```
